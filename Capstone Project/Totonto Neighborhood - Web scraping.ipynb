```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

response = requests.get(url="https://en.wikipedia.org/wiki/List_of_neighbourhoods_in_Toronto",)
print(response.status_code)
```

The status code code 200 shows everything is good

```python
wiki_page_text = response.text
```

```python
soup = BeautifulSoup(wiki_page_text, 'html.parser')
```

```python
#This returns a list containing all the tables in the HTML
table_soup = soup.find_all('table',class_="wikitable sortable")
#How many tables are there in this HTML?
#len(soup.find_all('table'))
table_soup
```

```python
# Search through the tables for the one with the headings we want.
for table in table_soup:
    ths = table.find_all('th')
    headings = [th.text.strip() for th in ths]
    if headings[:5] == ['CDN number', 'City-designated neighbourhood', 'Former city/borough', 'Neighbourhoods covered', 'Map']:
        break
table
```
